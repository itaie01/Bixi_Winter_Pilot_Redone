```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_22 = pd.read_csv('./Data/DonneesOuvertes2022.csv', sep=',')
data_23 = pd.read_csv('./Data/DonneesOuvertes2023.csv', sep=',')
```

```python
start_station_counts_22 = data_22['STARTSTATIONNAME'].value_counts().reset_index()
start_station_counts_22.columns = ['station_name', 'count']
start_station_counts_22.head()
```

```python
end_station_counts_22 = data_22['ENDSTATIONNAME'].value_counts().reset_index()
end_station_counts_22.columns = ['station_name', 'count']
end_station_counts_22.head()
```

```python
start_station_counts_23 = data_23['STARTSTATIONNAME'].value_counts().reset_index()
start_station_counts_23.columns = ['station_name', 'count']
start_station_counts_23.head()
```

```python
end_station_counts_23 = data_23['ENDSTATIONNAME'].value_counts().reset_index()
end_station_counts_23.columns = ['station_name', 'count']
end_station_counts_23.head()
```

```python
start_station_merged = start_station_counts_22.merge(start_station_counts_23, on='station_name', how='inner')
start_station_merged.columns = ['station_name', 'count_22', 'count_23']
start_station_merged['total_count'] = start_station_merged['count_22'] + start_station_merged['count_23']
start_station_merged.head()
```

```python
end_station_merged = end_station_counts_22.merge(end_station_counts_23, on='station_name', how='inner')
end_station_merged.columns = ['station_name', 'count_22', 'count_23']
end_station_merged['total_count'] = end_station_merged['count_22'] + end_station_merged['count_23']
end_station_merged.head()
```

## 150 Most Popular Stations over the course of the 2022 - 2023 BIXI seasons

```python
all_merged = start_station_merged.merge(end_station_merged, on='station_name', how='inner')
all_merged['total_combined_count'] = all_merged['total_count_x'] + all_merged['total_count_y']
all_merged = all_merged[['station_name', 'total_combined_count']]
all_merged = all_merged.sort_values(by='total_combined_count', ascending=False)
all_merged.columns = ['STARTSTATIONNAME', 'total_combined_count']
all_merged = all_merged.head(n=150)
all_merged.head()
```

## 150 Most Popular Station Coordinates

```python
station_locations = all_merged.merge(data_23[['STARTSTATIONNAME', 'STARTSTATIONLATITUDE', 'STARTSTATIONLONGITUDE']], on='STARTSTATIONNAME', how='inner')
station_locations = station_locations.drop_duplicates()
station_locations.head()
```

```python
print(*station_locations['STARTSTATIONLATITUDE'])
```

```python
lat_lon_pairs = [(lat, lon) for lat, lon in zip(station_locations['STARTSTATIONLATITUDE'], station_locations['STARTSTATIONLONGITUDE'])]
print(lat_lon_pairs)
```

```python
with open('./150_most_pop_station_coords.txt', 'w') as f:
    for lat_lon in lat_lon_pairs:
        f.write(f"{lat_lon}\n")

f.close()
```

```python
with open('./150_most_pop_station_coords.txt', 'r') as f:
    for line in f.readlines():
        line = line.strip()
        lat, lon = line.strip()[1:len(line)-1].split(',')
        lat_lon_pairs.append((float(lat), float(lon)))

f.close()
print(lat_lon_pairs)
```

```python
from qgis.core import QgsFeature, QgsField, QgsFields, QgsVectorLayer, QgsProject, QgsPointXY, QgsCoordinateReferenceSystem

lat_lon_pairs = []
with open('/Users/itai-epstein/Documents/Development/bixi_winter_pilot/150_most_pop_station_coords.txt', 'r') as f:
    for line in f.readlines():
        line = line.strip()
        x = line.strip()[1:len(line)-1].split(',')
        lat_lon_pairs.append((float(x[0]), float(x[1])))

# Define the desired coordinate reference system (CRS) with geographic (lat/long) coordinates
desired_crs = QgsCoordinateReferenceSystem('EPSG:4326')  # WGS 84

# Create a memory layer for points
point_layer = QgsVectorLayer('Point?crs=' + desired_crs.authid(), 'PointLayer', 'memory')

# Define fields for the layer
fields = QgsFields()
fields.append(QgsField('Latitude', QVariant.Double))
fields.append(QgsField('Longitude', QVariant.Double))
point_layer.dataProvider().addAttributes(fields)
point_layer.updateFields()

# Set the coordinate system for the layer
point_layer.setCrs(desired_crs)

# Create points and add them to the layer
features = []
for lat, lon in lat_lon_pairs:
    point = QgsPointXY(lon, lat)
    feature = QgsFeature()
    feature.setGeometry(QgsGeometry.fromPointXY(point))
    feature.setAttributes([lat, lon])
    features.append(feature)

point_layer.dataProvider().addFeatures(features)

# Add the layer to the map
QgsProject.instance().addMapLayer(point_layer)

# Refresh the map canvas
iface.mapCanvas().refresh()
```
